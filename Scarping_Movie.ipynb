```python
from bs4 import BeautifulSoup #Importing necessary Library
import requests
import pandas as pd
import numpy as np
```

```python
list1 = []  # Appending the values to this list
for page in range(1, 6, 1):
    url = f"https://subslikescript.com/movies?page={page}"  # websites that are going to be scraped
    result = requests.get(url)
    content = result.text  # Converting the URL to text
    soup = BeautifulSoup(content, 'html.parser')

    # Check if the 'main-article' class is present before proceeding
    box = soup.find('article', class_="main-article")
    if box:
        box1 = box.find_all('a', href=True)  # by inspecting the elements from websites

        for each in box1:
            list1.append(each.text)
```

```python
list1 #This is the appended values
```

```python
# Initialize empty lists to store names and years
names = []
years = []
# Iterate through the strings and split them
for string in list1:
    # Split based on parentheses
    parts = string.split(' (')
    if len(parts) == 2:
        # Extract the name and year
        name = parts[0]
        year = parts[1].strip(')')
        
        # Append to the respective lists
        names.append(name)
        years.append(year)

# Create a DataFrame
data = {'Name': names, 'Year': years}
df = pd.DataFrame(data)

# Print the DataFrame
df.head()
```
